```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/List_of_universities_and_colleges_in_Kenya"
page = requests.get(url)
soup = BeautifulSoup(page.text,"html")
```

```python
print(soup)
```

```python
soup.find_all("table")[1]
```

```python
table = soup.find_all("table")[1]
tHeaders = table.find_all("th")
print(tHeaders)
```

```python
tHeaders_titles = [title.text.strip() for title in tHeaders]
tHeaders_titles[0] = "NO"
print(tHeaders_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = tHeaders_titles)
```

```python
column_data = table.find_all("tr")

for row in column_data[1:]:
  row_data = row.find_all("td")
  individual_row_data = [data.text.strip() for data in row_data]

  if len(individual_row_data) < 4:
        individual_row_data.append('')  # Pad with empty string if fewer than 4 columns

  #print(individual_row_data)


  length = len(df)
  df.loc[length] = individual_row_data
```

```python
# Export DataFrame to CSV
df.to_csv('universities.csv', index=False)
```
